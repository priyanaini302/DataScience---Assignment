```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io  as pio
pio.renderers.default="svg"
```

```python
dataset=pd.read_csv("Zomato_Mumbai_Dataset.csv",delimiter='|')
```

```python
dataset.head
```

```python
dataset.shape
```

```python
dataset.info
```

```python
dataset.describe()
```

```python
wr=dataset['PAGE NO']=='PAGE NO'
```

```python
dataset[wr]
```

```python
datas=dataset[~wr]
```

```python
datas.drop(['URL','PAGE NO','CITY'],axis=1,inplace=True)
```

```python
datas.head
```

```python
datas.isnull().sum()
```

```python
datas[datas['PRICE'].isnull()]
```

```python
datas=datas.drop(labels=15080,axis=0)
```

```python
datas.fillna("NA",inplace=True)
```

```python
datas.isnull().sum()
```

```python
datas['RATING'].value_counts()
```

```python
datas['RATING'].replace(to_replace=['-','NEW','Opening'],value=0,inplace=True   )
```

```python
datas['PRICE']=datas['PRICE'].astype('int64')
datas['RATING']=datas['RATING'].astype('float64')
```

```python
datas.info()
```

```python
datas['TIMING'].value_counts()
```

```python
temp=datas["TIMING"].str.split("(",n = 1, expand = True)
temp
```

```python
datas['TIMING']=temp[0]
datas['DAYS_OPEN']=temp[1]
datas.head()
```

```python
datas['DAYS_OPEN']=datas['DAYS_OPEN'].str.replace(')','',regex=True)
datas.head()
```

```python
datas.isnull().sum()
```

```python
datas.fillna("NA",inplace=True)
datas.info()
```

```python
uselessdata=(datas['RATING']==0.0) | (datas['VOTES']==0)
datas[uselessdata]
```

```python
datas=datas[~uselessdata]
```

```python
datas['RATING_TYPE'].value_counts()
```

```python
datas['RATING_TYPE'].replace(to_replace='Excelente', value='Excellent', inplace=True)
datas['RATING_TYPE'].replace(to_replace=['Veľmi dobré', 'Bardzo dobrze', 'Muy Bueno', 'Velmi dobré'],value='Very Good',inplace=True)
datas['RATING_TYPE'].replace(to_replace=['Skvělá volba', 'Dobrze', 'Bueno', 'Buono', 'Dobré', 'Bom', 'Skvělé'], value='Good',inplace=True)
datas['RATING_TYPE'].replace(to_replace=['Priemer', 'Média', 'Çok iyi'], value='Average', inplace=True)
datas['RATING_TYPE'].replace(to_replace=["Průměr", 'Promedio', 'Ortalama', 'Muito Bom', 'İyi'], value='Poor', inplace=True)
datas['RATING_TYPE'].replace(to_replace=["Baik", "Biasa", 'Média ', "Sangat Baik"], value='Very Poor', inplace=True)
```

```python
datas['RATING_TYPE'].value_counts()
```

```python
datas['REGION'].value_counts()
```

```python
datas['REGION'].value_counts()
```

```python
datas['REGION'] = datas['REGION'].str.replace('[a-zA-Z].+--','',regex=True)
```

```python
datas["REGION"]=datas["REGION"].str.replace('West|west|East|east','',regex=True)
```

```python
datas["REGION"].value_counts()
```

```python
datas['REGION']=datas['REGION'].str.replace('Bandra Kurla Comples', 'Bandra', regex=True)
datas['REGION']=datas['REGION'].str.replace("CHD-Belapur","CBD Belapur", regex=True)
datas['REGION']=datas['REGION'].str.replace('Girgaon Chowpatty', 'Chowpatty', regex=True)
datas['REGION']=datas['REGION'].str.replace('Dadar Shivaji Park', 'Dadar', regex=True)
datas['REGION']=datas['REGION'].str.replace("Flea Bazaar Café Kanala Mills Compound", "Lower Parel", regex=True)
datas['REGION']=datas['REGION'].str.replace('Runwal Green', "Mulund", regex=True)
datas['REGION']=datas['REGION'].str.replace('Mumbai CST Area', 'Mumbai Central', regex=True)
datas['REGION']=datas['REGION'].str.replace("Kopar Khairane Seawoods Turbhe Ulme", "Navi Mumbai", regex=True)
datas['REGION']=datas['REGION'].str.replace('New Panvel Old Panvel', 'Panvel',regex=True)
datas['REGION']=datas['REGION'].str.replace("Kanothe", 'Sion', regex=True)
datas['REGION']=datas['REGION'].str.replace("Ghodbunder Road Majiwada", "Thane", regex=True)
```

```python
datas.head()

```

```python
datas.info()
```

```python
datas[datas.duplicated()]
```

```python
datas=datas.drop_duplicates()
```

```python
zomato=datas.copy()
```

# Perncentage of retaurants by rating

```python
rating=zomato['RATING_TYPE'].value_counts().reset_index()
rating.rename(columns={'index':'RATING TYPE','RATING_TYPE':'COUNT OF RETAURANTS'},inplace=True)
rating
```

```python
fig=px.pie(rating,names="RATING TYPE",values="COUNT OF RETAURANTS",
           color='RATING TYPE',title='percentage of restaurants by rating').update_traces(textposition='inside',
            textinfo='percent+label')
fig.show()

```

# Seafood restaurants

```python
seafood=zomato[zomato['CUSINE_CATEGORY'].str.contains('Seafood')]
seafood.sort_values(by='RATING',ascending=False).head(10)
```

```python
highest=zomato[zomato['RATING']>=4.5]
highest
```

```python
fig=px.histogram(highest,x='REGION',color='CUSINE TYPE',
                title='No.of best retaurants for each cusine type by places').update_xaxes(categoryorder="total descending")
fig.show()
```

# avg price rated restaurant

```python
highest_price=highest.groupby(by=['REGION','CUSINE TYPE'])['PRICE'].mean().reset_index()
highest_price
```

```python
fig=px.scatter(highest_price,x='REGION',y="PRICE",color="CUSINE TYPE",
              title='avg price distribution of high rated restaurant for each cuisine').update_traces(marker_size=10)
fig.show()
```

```python
max_price=zomato['PRICE'].max()
one_fourth_price=max_price/4
one_fourth_price
```

```python
aff_res=zomato[['NAME','PRICE','CUSINE_CATEGORY','REGION']]
aff_res=aff_res[aff_res['PRICE']<=1250]
aff_res.sort_values(by='PRICE',inplace=True)
aff_res
```

```python
highrate_res=highrate_res=zomato[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','RATING']]
highrate_res=highrate_res[highrate_res['RATING']>=4.5]
highrate_res.sort_values(by='PRICE',inplace=True)
highrate_res
```

```python
highrate_aff_df = pd.merge(aff_res, highrate_res, how='inner', on=['NAME', 'REGION'])
highrate_aff_df=highrate_aff_df[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE']]
highrate_aff_df.rename(columns={'NAME':"NAME", 'PRICE_x':"PRICE", "CUSINE CATEGORY_x":"CUSINE_CATEGORY",
            'REGION':"REGION", 'CUSINE TYPE':'CUSINE TYPE'},inplace=True)
```
